\end{array}$

$\left|x_{n+1}-\sqrt{a}\right|\leq C\left|x_{n}-\sqrt{a}\right|^{2}.$

$3.a. 3.b.$ 
Fórmula en general. Consideremos un círculo de radio $r$ y $P_n$ un poligono de  $n$ lados.
Nos interesa obtener el perimetro de poligonos $P_n$ inscritos en el circulo de radio $r$ ya que sabemos que al hacer $n$ suficientemente grande el poligono se aproxima  a la circunferencia y por lo tanto el perimetro del poligono se aproxima al valor $2\pi$. De este modo $\pi\approx \frac{\text{perímetro}}{2}$.
 
Comenzamos por dividir al polígono en n triángulos del mismo tamaño.


```python
from IPython.core.display import Image
Image(filename='Tarea1_Poligono.png')
```

Observamos que el segmento que va del centro de la circunferencia a cualquier vértice del polígono mide $r$. Sabemos también que la suma de los ángulos que se forman en el centro suman 360° por lo que para cada triángulo se tiene $\theta = 360/n$.

Bisectando cada ángulo central formamos triángulos rectángulos. 

Podemos encontrar el valor de x utilizando la identidad  $\sin(\frac{\theta}{2})=\frac{x}{r} \text{ de donde } x=r sin(\frac{\theta}{2})$. Se observa entonces que cada lado de $P_n$ tiene longitud $2rsin(\theta)$ y por lo tanto el perímetro de P está dado por la fórmula:

$Perímetro=2r\sin(\frac{\theta}{2})n$.

Finalmente, podemos aproximar $\pi\approx r\sin(\frac{\theta}{2})n $. 

```python
#Implementación
from numpy import sin,pi
def aprox_pii(n,r=None):
    if r is None:
        r=1
    theta=(360./n)*pi/180. #convertir a radianes
    perim= 2*r*sin(theta/2.)*n
    return perim/2.
```

Ahora podemos calcular el perimetro de poligonos de n lados y con ellos dar diferentes aproximaciones de $\pi$. Por ejemplo, para $P_n$ con $n=4,5,6$  se tienen las aproximaciones:

```python
aprox_pii(4)
```

```python
aprox_pii(5)
```

```python
aprox_pii(6)
```

Sin embargo ahora es necesario considerar la relación $\tan(\theta/2)=\frac{x}{r}$ de donde $x=r\tan(\frac{\theta}{2})$. En este caso aproximamos como sigue: $\pi \approx 2r\tan(\frac{\theta}{2})n$.


```python
##Implementación
from numpy import tan
def aprox_pic(n,r=None):
    if r is None:
        r=1
    theta=theta=(360./n)*pi/180. #convertir a radianes
    perim= 2*r*tan(theta/2.)*n
    return perim/2.
```
